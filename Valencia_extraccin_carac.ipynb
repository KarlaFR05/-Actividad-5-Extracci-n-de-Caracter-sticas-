# **host_response_rate**

Determinación de clases para datos agrupados

```python
#Calculamos el numero total de la poblacion n
Valencia['host_response_rate'].info()
n=9447
```

```python
# Convertir la columna de porcentaje de str a numérico (decimal)
Valencia['host_response_rate'] = Valencia['host_response_rate'].replace('%', '', regex=True).astype(float) / 100
```

```python
#Obtenemos el límite superior y el límite inferior de la columna objetivo
Max = Valencia['host_response_rate'].max()
Min = Valencia['host_response_rate'].min()

Limites = [Min, Max]
Limites
```

```python
#Calculamos el Rango R
R = Max - Min
R
```

```python
#Calculamos el número de Intervalos de Clase "ni", aplicando la regla de Sturges
ni = 1+3.32*np.log10(n)
ni
```

```python
#Calculamos el Ancho del intervalo "i"
i=R/ni
i
```

CREACIÓN DE LAS CATEGORÍAS

```python
#Categorización de variables
intervalos = np.linspace(Min, Max, 9)
intervalos
```

```python
#Creamos las categorías
categorias = ["Categoria 1", "Categoria 2", "Categoria 3", "Categoria 4",
              "Categoria 5", "Categoria 6", "Categoria 7", "Categoria 8"]
```

```python
#Finalmente creamos las categorías en la columna numérica
Valencia['host_response_rate'] = pd.cut(x=Valencia['host_response_rate'], bins=intervalos, labels= categorias)
Valencia['host_response_rate']
```
